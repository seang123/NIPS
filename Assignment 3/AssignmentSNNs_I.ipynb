## The leaky integrate and fire (LIF) neuron

First, construct a leaky integrate and fire neuron which follows the dynamics:
$$
\tau_m \frac{dv}{dt} = (v_{\text{rest}} - v(t)) + I(t)
$$

With the following condition for spiking
$$
v(t) = 
\begin{cases}
    v(t) & \text{if } v(t) < v_{thresh} \\
    v_{rest} & \text{otherwise, record a spike and reset}
  \end{cases}
$$

Analytical solution assuming $I(t)$ is a constant $I_0$

$$
v(t) = (v_{rest} + RI_0) - ((v_{rest} + RI_0) - v_0)exp(\frac{-t}{\tau_m})
$$

Plot the membrane voltage evolution over time of a single neuron (parameters given below). Also plot the final spike-train. **(2 points)**

```python
# Parameters
v_rest = -50.0      #mV
v_thresh = -25.0    #mV
tauM = 20.0         #ms
I = 30.0            #mV
timestep = 0.1      #ms

# Simulate for 200ms
total_time = 200.0
nb_timesteps = int(total_time / timestep)
```

```python
v_act = np.zeros(nb_timesteps) # membrane potential across time.
spike_train = np.zeros(nb_timesteps)

# initilise the membrane potential to its rest potential.
v_old = v_rest

for i in range(nb_timesteps):
    
    # update the membrane potential
    v_old = (v_rest + I) - ((v_rest + I) - v_old) * np.exp(-timestep/tauM)

    # if mem. potential is > threshold reset it
    if v_old >= v_thresh:
        v_old = v_rest
        spike_train[i] = 1
        
    # save the membrane potential for plotting
    v_act[i] = v_old
        
# plt.figure(0)
fig, axs = plt.subplots(2, 1, figsize=(15, 5))

axs[0].plot(v_act)  
axs[0].set_title("Membrane Potential")
# axs[0].set_xlabel("time")
axs[0].set_ylabel("voltage")

# plt.figure(1)
axs[1].plot(spike_train)
axs[1].set_title("Spike train")
axs[1].set_xlabel("time step (0.1ms)")
axs[1].set_ylabel("")

plt.subplots_adjust(hspace=0.3)

x = np.nonzero(spike_train)
print(x)
```

### The input-output curve of a LIF neuron

If there is a fixed input ($I$) to a LIF neuron it is possible to compute an analytic solution to this equation. Derive the analytic solution and plot the time to spike (from rest) of a LIF neuron across a range of inputs (range given below). **(2 point)**

$$
v(t) = (v_{rest} + RI_0) - ((v_{rest} + RI_0) - v_0)exp(\frac{-t}{\tau_m})
$$

```python
I_range = np.linspace(0,100,int(1e5))

# set v_o = v_rest
# rearange for t = ...
# set v(t) = v_thresh  <-- because we want the time to spike
# see how the solution varies across different I

v_0 = v_rest
v_t = v_thresh

times = np.zeros(np.size(I_range))

for i in range(0, np.size(I_range)):
    
    t = -tauM * np.log( ((v_rest + I_range[i]) - v_t ) / (v_rest + I_range[i] - v_0) )
    times[i] = t
    
    
plt.figure(1)
plt.plot(I_range, times)
plt.title("Spike times for different input currents")
plt.ylabel("time to spike (ms)")
plt.xlabel("input current (mV)")
plt.show()
```

Now plot the firing rate of a LIF neuron across this range (inverse of the time to spike).  **(1 point)**

```python
v_0 = v_rest
v_t = v_thresh 

times = np.zeros(100000)

for i in range(0, np.size(I_range)):
    
    t = 1 / ( -tauM * np.log( ((v_rest + I_range[i]) - v_t ) / (v_rest + I_range[i] - v_0) ) )
    times[i] = t
    
# convert the nan's to 0's
times = np.nan_to_num(times)
    
plt.figure(1)
plt.plot(I_range, times)
plt.title("Firing rates for different input currents")
plt.ylabel("firing rate (Hz)")
plt.xlabel("input current (mV)")
plt.show()

```

## The Hodgkin Huxley Model

Now that you have simulated a simple spiking neuron. Let us also re-produce the dynamics of a more complex neuron with the Hodgkin Huxley (HH) model.


https://www.st-andrews.ac.uk/~wjh/hh_model_intro/

The dynamics of the HH neuron are described

$$
C_m\frac{dv}{dt}= g_K \; n^4 \; (E_K - v) + g_{Na} \; m^3 \; h \; (E_{Na} - v) + g_L \;(E_L - v) + I
$$

where

$$
\frac{dn}{dt} = \alpha_n(1-n) - \beta_n n \\
\frac{dm}{dt} = \alpha_m(1-m) - \beta_m m \\
\frac{dh}{dt} = \alpha_h(1-h) - \beta_h h \\
$$

and the equations for the $\alpha$s and $\beta$s are given:
Screen Shot 2020-09-14 at 11.38.57.png

Note, to avoid strange initial dynamics you should initialize v, n, m, and h such that:
$$
v(0) = -65.0 \\
n(0) = \alpha_n(-65.0) / (\alpha_n(-65.0) + \beta_n(-65.0)) \\
m(0) = \alpha_m(-65.0) / (\alpha_m(-65.0) + \beta_m(-65.0)) \\
h(0) = \alpha_h(-65.0) / (\alpha_h(-65.0) + \beta_h(-65.0))
$$



Details extracted from the [Neuronal Dynamics Textbook](https://neuronaldynamics.epfl.ch/online/Ch2.S2.html), Chapter 2.2 (Table Above Equation 2.6).


$$
\frac{dn}{dt} = \frac{0.02*(u - 25)}{1-exp(-\frac{u-25}{9})}*(1-n) - \frac{-0.002(u-25)}{1-exp(\frac{u-25}{9})} * n
$$


```python
EL, gL = -65.0, 0.3
ENa, gNa = 55.0, 40.0
EK, gK = -77.0, 35.0
C = 1.0

timestep = 0.01
total_time = 500.0
nb_timesteps = int(total_time / timestep)

def get_alphas(u):
    """
    Alphas calculation for N, M, and H
    
    return:
        n, m, h
    """
    n = (0.02 * ( u - 25)) / (1-np.exp(-(u-25)/9))
    m = (0.182 * (u + 35)) / (1-np.exp(-(u+35)/9))
    h = 0.25*np.exp(-(u+90)/12)
    return n, m, h

def get_betas(u):
    """
    Betas calculation for N, M, and H
    
    return:
        n,m,h
    """
    n = (-0.002* (u - 25)) / (1-np.exp((u-25)/9))
    m = (-0.124 * (u + 35)) / (1-np.exp((u+35)/9))
    h = (0.25*np.exp((u+62)/6))/np.exp((u+90)/12)
    return n, m, h              

def calc_nmh(v):
    """
    Uses alphas and betas to calculate N, M, H
    """
    alphas = get_alphas(v)
    betas = get_betas(v)
    n = alphas[0]/( alphas[0] + betas[0] )
    m = alphas[1]/( alphas[1] + betas[1] )
    h = alphas[2]/( alphas[2] + betas[2] )
    return n, m, h

def dn_dm_dh(nmh, alphas, betas):
    n, m, h = nmh[0], nmh[1], nmh[2]
    
    n = alphas[0] * (1-n) - betas[0] * n
    m = alphas[1] * (1-m) - betas[1] * m
    h = alphas[2] * (1-h) - betas[2] * h
    
    return n, m, h
    
def dV_dt_HH(EL, gL, ENa, gNa, EK, gK, C, nmh, v_t, I):
    """
    I : input potential
    v_t: current membrane potential
    
    returns change in membrane potential in one time step
    """
    n, m, h = nmh[0], nmh[1], nmh[2]
    
    return (gK * n**4 *(EK - v_t) + gNa * m**3 * h * (ENa - v_t) + gL * (EL - v_t) + I) / C


def hodgkin_huxley(EL, gL, ENa, gNa, EK, gK, C, timestep, total_time, I=1):

    v_t = -65
    nmh = calc_nmh(v_t)
    n, m, h = calc_nmh(v_t)

    times = np.zeros(shape = (nb_timesteps))
    ns = np.zeros(shape = (nb_timesteps))

    for t in range(0, nb_timesteps):
        times[t] = v_t
        ns[t] = n

        #update v_t: the voltage across the membrane at each time step
        v_t = v_t + timestep * dV_dt_HH(EL, gL, ENa, gNa, EK, gK, C, [n,m,h], v_t, I)

        #get the alpha and beta values for n, m, h at each new time step
        alphas = get_alphas(v_t)
        betas = get_betas(v_t)
        
        # get the change in n, m, h values
        derivatives = dn_dm_dh([n, m, h], alphas, betas)

        #update n,m,h based on: n,m,h + timestep*derivatve n,m,h
        n = n + timestep * derivatives[0]
        m = m + timestep * derivatives[1]
        h = h + timestep * derivatives[2]


    return np.array(times), np.array(ns)
 
```

Plot the voltage of a neuron with:

- sub-threshold input
- above threshold input (producing a single spike)
- above threshold input producing repeated spiking

 **(4 points)**


```python
fig, axs = plt.subplots(3, figsize=(10,10))
sub_threshold, _ = hodgkin_huxley(EL, gL, ENa, gNa, EK, gK, C, timestep, total_time, 0.2)

# plot the sub_threshold spike
axs[0].plot(np.arange(0, total_time, timestep), sub_threshold)
axs[0].set_xlabel("time")
axs[0].set_ylabel("Euler approx. membrane voltage (mV)")
axs[0].set_title("Input current: 0.2")

# plot the above threshold - single spike
above_thresh_single, _ = hodgkin_huxley(EL, gL, ENa, gNa, EK, gK, C, timestep, total_time, 0.3)
axs[1].plot(np.arange(0, total_time, timestep), above_thresh_single )
axs[1].set_xlabel("time")
axs[1].set_title("Input current: 0.3")

# plot the above threshold - continuous spiking
repeating, _ = hodgkin_huxley(EL, gL, ENa, gNa, EK, gK, C, timestep, total_time, 0.6)
axs[2].plot(np.arange(0, total_time, timestep), repeating)
axs[2].set_xlabel("time")
axs[2].set_title("Input current: 0.6")

plt.subplots_adjust(hspace=0.3)
plt.show()
```
